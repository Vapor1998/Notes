The Decision Tree is dervied from the Table of three Columns **College, Projects, Intern? Call?**  
In this, **College** has the values Tier1, Tier2 and Tier3. **Projects and INtern?** has Yes and No **Call?** also has Value Yes and No and this is the OUTPUT column.  

Now while making Tree, we first Check whether College is Tier1 and Tier2, if it is Tier1 and Tier2, We will further check if the Individual has internship or not, if he has internship, he will get the call. If he does not have the Internship, we will check if he has projects or not, if he has project, he will get call(Meaning output Yes) else he will get output No.  
  
Coming to First part if the college is not Tier1 and Tier2, we will check if the individual has internship or not, if he does not have, he will not get the call meaning output No. If he has Internship, we will further check if the individual has projects or not, if he has he will get the call else he will not get the call.
