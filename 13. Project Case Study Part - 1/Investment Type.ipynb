```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("startup_funding.csv")

df = data.copy()

investment =df["InvestmentType"].value_counts()

amount = df["AmountInUSD"].str.replace(',','')
amount.fillna(0, inplace = True)

df["InvestmentType"] = df["InvestmentType"].replace(["SeedFunding",'PrivateEquity', 'Crowd funding'], ["Seed Funding",'Private Equity', 'Crowd Funding'])

investments = df["InvestmentType"]

df1 = pd.DataFrame(investments)
df2 = pd.DataFrame(amount)
df2["InvestmentType"] = investments

invest = df['InvestmentType'].value_counts().index

invest=np.array(invest)
amt = np.array(amount, dtype = int)

arr = []

for i in invest:
    ind = np.where(investments == i)
    sums = amt[ind].sum()
    arr.append([i, sums])

    
arr.sort(reverse = True)

invt = []
usd = []
for i in arr:
    invt.append(i[0])
for i in arr:
    usd.append(i[1])
    
plt.pie(usd, autopct = '%.2f%%', labels = invt)
plt.show()

arr2 = []
for i in range(len(usd)):
    per = usd[i] / sum(usd) * 100
    arr2.append([invt[i], format(per, '.2f')])

arr2[0],arr2[1] = arr2[1], arr2[0]
for i in arr2:
    for j in i:
        print(j, end= ' ')
    print()
                
```

```python

## correct as well


import pandas as pd
import matplotlib.pyplot as plt

def correctValue(x):
    if x=='PrivateEquity':
        return 'Private Equity'
    elif x=='SeedFunding':
        return 'Seed Funding'
    elif x=='Crowd funding':
        return 'Crowd Funding'
    else:
        return x
    
df = pd.read_csv('startup_funding.csv')
df.dropna(inplace=True,subset=["InvestmentType"])
df["InvestmentType"] = df["InvestmentType"].apply(lambda x :correctValue(x))
ans=[]
values = {"AmountInUSD": 0}
df.fillna(value=values,inplace=True)

df["AmountInUSD"]=df["AmountInUSD"].apply(lambda x:int("".join(str(x).split(","))))

for ele in df["InvestmentType"].unique():
    ans.append([ele,df[df["InvestmentType"]==ele].AmountInUSD.sum()])


ans.sort(key=lambda x:x[1],reverse=True)
#ans[2],ans[3]=ans[3],ans[2]
s = sum([ans[i][1] for i in range(len(ans))])
plt.pie([round(ans[i][1]/s,2) for i in range(len(ans))])
plt.show()
for i in range(len(ans)):
    print(ans[i][0],format(ans[i][1]/s*100,".2f"))



    
```
