```python
import pandas as pd
import numpy as np
columns = ['sl', 'sw', 'pl',' pw','variety']

iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names = columns)
df = iris.copy()
a = df[df['variety'] == 'Iris-setosa'].describe()
b = df[df['variety'] == 'Iris-versicolor'].describe()
c = df[df['variety'] == 'Iris-virginica'].describe()

d = [a,b,c]

x = ['Iris-setosa','Iris-versicolor', 'Iris-virginica'  ]
s = 0
for i in d:
    for j in i.loc['min']:
        print(format(j,'.2f') , end = ' ')
    print(x[s])
    for k in i.loc['max']:
        print(format(k,'.2f') , end = ' ')
    print(x[s])
    for l in i.loc['mean']:
        print(format(l,'.2f') , end = ' ')
    print(x[s])
    
    s+=1
    

```

```python
## Open and read data file as specified in the question
## Print the required output in given format
import pandas as pd 

columns = ['sepal.length', 'sepal.width', 'petal.length','petal.width','variety']

data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names = columns)

#data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
#print(data)

names=list(set(data['variety']))
#print(names)
#l=set(data[])
ise=data[data['variety']==('Iris-setosa')]
ive=data[data['variety']==('Iris-versicolor')]
ivs=data[data['variety']==('Iris-virginica')]


print (format(min(ise['sepal.length']), '.2f'),
       format(min(ise['sepal.width']), '.2f'),
       format(min(ise['petal.length']), '.2f'),
       format(min(ise['petal.width']), '.2f'),
       'Iris-setosa')

print (format(max(ise['sepal.length']), '.2f'),
       format(max(ise['sepal.width']), '.2f'),
       format(max(ise['petal.length']), '.2f'),
       format(max(ise['petal.width']), '.2f'),
       'Iris-setosa')

print (format((ise['sepal.length'].mean()), '.2f'),
       format((ise['sepal.width'].mean()), '.2f'),
       format((ise['petal.length'].mean()), '.2f'),
       format((ise['petal.width'].mean()), '.2f'),
       'Iris-setosa')

print (format(min(ive['sepal.length']), '.2f'),
       format(min(ive['sepal.width']), '.2f'),
       format(min(ive['petal.length']), '.2f'),
       format(min(ive['petal.width']), '.2f'),
       'Iris-versicolor')

print (format(max(ive['sepal.length']), '.2f'),
       format(max(ive['sepal.width']), '.2f'),
       format(max(ive['petal.length']), '.2f'),
       format(max(ive['petal.width']), '.2f'),
      'Iris-versicolor')

print (format((ive['sepal.length'].mean()), '.2f'),
       format((ive['sepal.width'].mean()), '.2f'),
       format((ive['petal.length'].mean()), '.2f'),
       format((ive['petal.width'].mean()), '.2f'),
       'Iris-versicolor')

print (format(min(ivs['sepal.length']), '.2f'),
       format(min(ivs['sepal.width']), '.2f'),
       format(min(ivs['petal.length']), '.2f'),
       format(min(ivs['petal.width']), '.2f'),
       'Iris-virginica')

print (format(max(ivs['sepal.length']), '.2f'),
       format(max(ivs['sepal.width']), '.2f'),
       format(max(ivs['petal.length']), '.2f'),
       format(max(ivs['petal.width']), '.2f'),
      'Iris-virginica')

print (format((ivs['sepal.length'].mean()), '.2f'),
       format((ivs['sepal.width'].mean()), '.2f'),
       format((ivs['petal.length'].mean()), '.2f'),
       format((ivs['petal.width'].mean()), '.2f'),
       'Iris-virginica')


```
