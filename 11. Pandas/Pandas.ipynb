```python
iris = df.copy()
iris.head()
```

```python
#To get the column names use columns attribute

iris.columns
```

```python
#To get the indexs of the dataframe, index attribute is used

iris.index
```

```python
# drop() function is used to delete a row from the dataset
iris.drop(0)
iris.head()
```

```python
#No changes have been made because drop function does not delete the rows from the original dataframe. 
#It creates a copy of the dataframe and returns the updated copy

#To make changes in the original dataframe we need to pass a inplace=True argument. This ensures that the changes are made 
#in the original dataset

iris.drop(0, inplace=True)
iris.head()
```

```python
#To delete more than one columns in one go, we can pass a list of row numbers

iris.drop([4,5,8], inplace=True)
iris.head(10)
```

```python
# index[i] is used to access the ith row and not the row whose row no is i

print(iris.index[0])


#Therefore iris.index[i] can also be passed to drop function to delete the ith row
iris.drop(iris.index[0], inplace=True)
iris.head(10)
```

```python
#to get the ith row from starting we use iloc

print(iris.head())
iris.iloc[0]
```

```python
#Similarly to access the ith label row, we use loc

print(iris.head())
iris.loc[2]
```

```python
#To add a row into a dataframe, 

iris.loc[0] = [1, 2, 3, 4, "Iris-setosa"]

#The entry will be added to the end
iris.tail()
```

```python
#To reset the indices, reset_index() function is used

print("Dataframe before reset index")
print(iris.head(10))
iris.reset_index(inplace=True)
print("\nDataframe after reset index")
print(iris.head(10))

#Note that in the above dataframe the indices before resetting index are added as a column in our updated dataframe
```

```python
# To not add the previous index values in the dataframe, drop=True argument in passed in the reset_index function

print("Dataframe before reset index")
print(iris.head(10))
iris.reset_index(drop=True, inplace=True)
print("\nDataframe after reset index")
print(iris.head(10))
```

```python
#Now check the indices of the updated dataframe
iris.index
```

```python
#Deleting columns is similar to deleting rows.
#We use drop function to delete whole columns with an extra argument axis=1
#By default value of axis is 0. That means, rows will be deleted by default 

iris.drop('sl', axis=1, inplace=True)
iris.head()
```

# Terror DeadliestAttack

```python
import numpy as np
import pandas as pd

terror =pd.read_csv("terrorismData.csv")
df = terror.copy()

killed = df['Killed'].max()
killedindex = df[df["Killed"] == killed]  ## by adding  'index.values' at the end
                                          ## will give index of the particular value 
    
country = killedindex["Country"].to_string(index=False)   ## This is the way to print the value from
                                                          ## dataframe without label or index value.
    
killed = killedindex['Killed'].to_string(index = False)
group = killedindex['Group'].to_string(index = False)

kill = float(killed)
kills = int(kill)
print(kills, country , group)
```

# Terror Frequency

```python
import numpy as np
import pandas as pd

terror =pd.read_csv("terrorismData.csv")
df = terror.copy()

df["Casualty"] = df["Killed"] + df["Wounded"]
casualty = df["Casualty"]

state1 = df[df["State"] == 'Jammu and Kashmir']
state2 =df[df["State"] == 'Odisha']
state3 = df[df["State"] == 'Jharkhand']
state4 = df[df["State"] == 'Andhra Pradesh']
state5 = df[df["State"] == 'Chhattisgarh']

year = len(set(df["Year"]))


casualty1 = state1['Casualty'].sum()
casualty2 = state2['Casualty'].sum()
casualty3 = state3['Casualty'].sum()
casualty4 = state4['Casualty'].sum()
casualty5 = state5['Casualty'].sum()

frequency1 = casualty1 / year

frequency2 = (casualty5 + casualty4 + casualty3 + casualty2 )/ (year)


arr = [frequency2 ,frequency1]
for i in arr:
    print(int(i), end= ' ')


```

# Terror Government

```python
import numpy as np
import pandas as pd

terror =pd.read_csv("terrorismData.csv")
df = terror.copy()

dateindex = df[(df["Country"] == "India") & (df["Year"] == 2014) & (df["Month"] == 5) & (df["Day"] == 26) ].index.values[0]
sliced_df = df.iloc[dateindex:]
country = sliced_df[sliced_df["Country"] == "India"]
date = country["Day"].count()
group = country["Group"].value_counts().index[1]
print(date, group)
```
