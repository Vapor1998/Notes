```python
import pandas as pd

iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
iris 

##df = iris    ## If we make a copy, internal data of the frame is shared by both variable,
               ## so, changes made in one variable will reflect in other variable or the main 
               ## dataframe.
        
df = iris.copy()  ## So we use copy() function.

```

```python
df.columns = ['sl','sw','pl','pw', 'Flower_type']    ## this will change the column name.
```

```python
df.head()
```

```python
## Drop by label
a = df.drop(3, inplace = True)   ## To delete a row.
                                 ## drop() takes a label, not the position
df.head()
```

```python
df.index
```

```python
## Drop by position

df.drop(df.index[0], inplace = True)
df.head()
```

```python
## to delete more than one row at same time

df.drop(df.index[[0,1]], inplace = True)   ## Here we pass more than one entries in one array inside
                                           ## index.Looking at upper outpput, it will drop label 4
                                           ## and 5.As we have executed it 2 times, it will drop 6 
                                           ## and 7 as well.
df.head()
```

```python
df.sl > 5    ## It will check wether the column value is > 5 or not, it will return column in bool
             ## format. THIS WILL RETURN FULL COLUMN.
```

```python
df[df.sl  > 5 ]  ## It will give only those rows, which are > 5.THIS WILL NOT RETURN FULL COLUMN.
```

```python
df[df.Flower_type == 'Iris-setosa']   ## It will return only those rows, whose value == "Iris-setosa".

c1 = df[df.Flower_type == 'Iris-setosa']    ## we can select all flower types in variable like this 
                                           ## and we can compare them using describe().
c1
```

```python
## iloc vs loc
print(df.head())
print(df.iloc[0])   ## This will give values at index 0 of rows.
##print(df.loc[0])    ## This will fetch values from label.It will give error as there is no label 0.
print(df.loc[8])    ## This will not give error as there is label 8 at position 0.
```

```python
## add a row
df.loc[0] = [1,2,3,4, 'Iris-setosa']    ## This will add a row at label 0.

df.tail()   ## this is as same as head, head gives first five elements from dataframe, while,
            ## tail gives data from last five elements from dataframe.
```

# Manipulating Data in DataFrame - 2

```python
## It will reset index or label value

df.reset_index()

## but as we can see it has added unwanted index column, so we can drop that using drop = True
```

```python
df.reset_index(drop = True)  ## As we can see, it has dropped index column.

## Another thing is it is not making changes in og dataframe, it is actually giving you new
## dataframe. TO make changes in og dataframe, we do inplace = True.
```

```python
df.reset_index(drop = True, inplace = True)
df   ## Now it has changed og df.

##
```

```python
df.index  ## Now as we can see, its a range index.
```

```python
## To add column or delete column
df.drop('sl')   ## it will give error as we have learned, it will check for row 'sl',
                ## as there is no row 'sl', it will give error.
    
```

```python
## to delete a column, we add axis = 1 as the index of sl column is 1 in every row.
## axis means to run it columnwise. So it will look for column = 'sl' this time.
df.drop('sl', axis = 1)
## as we can see, column 'sl' is deleted.
## but it has not changed og df, it has returned new dataframe.
## to make changes in og df, we use inplace = True
```

```python
## we have already compiled the code with inplace = True and now if we run df, it will show 
## df with no sl column.
df
```

```python
## another way to delete a column is to call del function
del df['sw']  ## this deletes from og df.
```

```python
df  ## as we can see, sw is deleted.
```
