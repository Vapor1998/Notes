#### *Now we want to clean this files. By cleaning, we means we need to remove stopowrds and do lemmatization.*

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import string
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))
punctuation = list(string.punctuation)
stops.update(punctuation)
```

*Created stopwords and lemmatizer above*

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

**Explaination of this Function**  
This function will get the words.  
We need to clean these words.  
For that we need to create new array.
1. For each word in words, we need to check if w.lower() is in stops or not? if not,
2. If not, we will find the pos_tag. Now we need to pass the word w as it is without doing w.lower().  
   Because if we lower it, it might change the meaning of the word.  
3. Once we have the part of speech, we want to lemmatize it. If we know, pos cannot be passed directly, because  
   it requires to be different. So we use the get_simple_pos() function above to get the pos and then within that 
   we extract the pos for which the problem is solved in above cell with pos_tag.  
4. Then we will have the clean word which we get from lemmatizer and then we append it in the array and if we want, we can lower 
   it and append it.

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

**Now we go through each documentation to clean words**  
Now in this we go through each document, category in documents we have, and then we want to add new tuple in the documents array where the first word in the document is the cleaned words from the function and second is the category.  

```python
documents = [(clean_review(document), category) for document, category in documents]
```

# 7. Building Feature Set(Lecture)  
In nltk, when we want to supply data to the classifier, the format will be dictionary with all the features and the category.  
So for 2000 documents, we will convert them into a format like this. So 2000 tuples in the array where each tuple will have two parts, the first part is a dictionary of features and the second part will be the category.  
So what are the features we will be using? We will be using the top words that are available in the whole dataset that we have.  
So the first thing we will be using is we will make an array which has all the words. We will find the frequency distribution and we will take the words with maximum frequency as our features. Once we have those, lets say 3000 words, which are your featureset, then for each document that we have, we will create this dictionary of whether it has the first word or not and so on.  

```python
## we need to realize that creating these features should not be done on the complete dataset.
## It has to be done with training data and not with testing data.  
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
a = [1,2]
b = [3,4]
a +=  b
a
```

```python
## lets create containing all the words.
all_words = []

for doc in training_documents:
    all_words += doc[0]  ## because doc is a tuple and its 0th element is the word array.
    
```

```python
import nltk
```

```python
## To find the frequency distribution, 
freq = nltk.FreqDist(all_words)  ## this is a freq distribution object
```

```python
## once we store it, we can use functions like most common
freq.most_common()  ## we can pass parameter about how much frequent words we want.
```

```python
## now to get 3000 most words, 
common = freq.most_common(3000) 

## Now this is the list of tuples so we wants words in list
features = [i[0] for i in common]
```

```python
features
## these are the 3000 words which are our features.
```

*Now we have our features what do we want to do now is for each document, we want to create dictionary where this dictionary will have for eg: lets say for word film, {'film' : True or not} It means whether it comes in the document or not.If it comes the value of key is True else False.*  
We will ceate function which takes the words, and we will create dictionary for each document, then we will go through each words in the top 3000 features we created, we need to check if this word comes in the particular document, if this word w is the part of the array words(which are mentioned in the function), then we need to put a true else we need to put a false.  
We need to create a set of words.(which are of particular document we pass).


```python
def get_feature_dict(words):
    current_features =  {}
    words_set = set(words)
    for w in features:  
        current_features[w] = w in words_set   ## this will return True if w is in words_set else False
    return current_features
```

```python
## we try this function for 0th document

get_feature_dict(training_documents[0][0])

## yes we are getting feature dict
```

```python
## to check whether we pass array of sets 
training_documents[0][0]

## yes we did
```

```python
training_data = [(get_feature_dict(doc), category) for doc, category in training_documents]
## now we are done with training documents, we are not doing it with testing documents,
```

```python
testing_data = [(get_feature_dict(doc), category) for doc, category in training_documents]
```
