```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

allPages = ['http://books.toscrape.com/catalogue/page-1.html',
            'http://books.toscrape.com/catalogue/page-2.html']

column_names = ['Title', 'Link', 'Price', 'Quantity in Stock']

baseUrl = 'http://books.toscrape.com/catalogue/'

bookdetails = []

for i in allPages:
    response = requests.get(i)
    data = BeautifulSoup(response.text,'html.parser')
    books = data.find_all(class_ = 'product_pod')
    
    for i in books:
        b = []
        b1Url = baseUrl + i.h3.a['href']
        res = requests.get(b1Url)
        d = BeautifulSoup(res.text, 'html.parser')
        title = d.h1.string
        price = d.find(class_ = 'price_color').string
        stock = d.find(class_ = 'instock availability').contents[-1].strip()
        q = re.search('\d+', stock).group()
        p = re.search('[\d.]+', price).group()
        q = int(q)
        p = float(p)
        b.append(title)
        b.append(b1Url)
        b.append(p)
        b.append(q)
        bookdetails.append(b)



for i in bookdetails:
    for j in i:
        print(j, end = ' ')
    print()

    
    
```
