```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import operator

response = requests.get('https://www.imdb.com/search/title/?title_type=feature&release_date=2010-01-01,2014-12-31&sort=num_votes,desc')
html = response.text
data = BeautifulSoup(html, 'html.parser')

prefix = 'https://www.imdb.com'

an = data.find_all(class_= 'lister-item mode-advanced')

d = {}
for i in an:
    t = i.h3.a.string
    y = i.h3.contents[5].string.strip()
    x = re.search('\d+', y).group()
    x = int(x)
    if x in d:
        continue
    else:
        d[x] = t
        
sortedDictionary = sorted(d.keys())
for i in sortedDictionary:
    print(d[i])

        
```
