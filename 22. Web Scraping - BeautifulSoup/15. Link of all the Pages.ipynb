```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://books.toscrape.com/')
html_data = response.text

data = BeautifulSoup(html_data, 'html.parser')

## As we extracted all the data of the books from first page, we now know that there are data of 1000 books
## in this website on different pages. We have to extract all of them.
## There are 50 pages present in this website.
## To extract data of all 50 pages, we need to send request to all the 50 pages.

## Now as we did the inspection, we saw that there is a class next and there is an a tag inside which has the link to the
## next page.
## The link is in href which is 'page-2.html' for page 2.
## The link for page 3 is 'page-3.html'. This is not the complete url. The prefix is 'https://books.url.com/'.
## Now the url for page 1 is 'https://books.url.com/', for page 2 it is 'https://books.url.com/catalogue/page-2.html'.
## And then for page 3 it is 'https://books.url.com/page-3.html'.
## Now as we see there is an exception for page 2.
## Now as we go back from page 2 to page 1, the link is 'https://books.url.com/catalogue/page-1.html'.
## Then if we move from this first page to second, the link is 'https://books.url.com/catalogue/page-2.html'
## Then to third from this page, link is 'https://books.url.com/page-3.html'.
## To solve this issue, we can simply make an if else statement for first two links or 
## we can make request to the first page of this link 'https://books.url.com/catalogue/page-1.html'
## And then extract data of the second page and then extract data from the second page and so on.

```

```python
all_urls = ['http://books.toscrape.com/catalogue/page-1.html']

currentpage = 'http://books.toscrape.com/catalogue/page-1.html'

baseurl = 'http://books.toscrape.com/catalogue/'
response = requests.get(currentpage)


while response.status_code == 200:
    data = BeautifulSoup(response.text, 'html.parser')
    nextpage = data.find(class_ = 'next')
    if nextpage is None:
        break
    nextpageurl = baseurl + nextpage.a['href']
    print(nextpageurl)
    all_urls.append(nextpageurl)
    currentpage = nextpageurl
    response = requests.get(currentpage)
    
```
