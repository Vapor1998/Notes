```python
link = 'https://quotes.toscrape.com/'

import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import operator

response = requests.get(link)

d = {}
while response.status_code == 200:
    html = response.text
    data = BeautifulSoup(html,'html.parser')
    au = data.find_all(class_ = 'author')
    for i in au:
        author = i.string
        if author in d:
            d[author] = d.get(author, 0) + 1
        else:
            d[author] = 1
        
    
    li = data.find(class_ = 'next')
    if li is None:
        break
    post = li.a['href']
    url = link+ post
    response = requests.get(url)

for i in d:
    print(i)
```

```python
link = 'https://quotes.toscrape.com/'

import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import operator

response = requests.get(link)

d = {}
while response.status_code == 200:
    html = response.text
    data = BeautifulSoup(html,'html.parser')
    au = data.find_all(class_ = 'author')
    for i in au:
        author = i.string
        if author in d:
            d[author] = d.get(author, 0) + 1
        else:
            d[author] = 1
        
    
    li = data.find(class_ = 'next')
    if li is None:
        break
    post = li.a['href']
    url = link+ post
    response = requests.get(url)

key = d.keys()
s = ''
for i in key:
    s = s+i+','
st = s.split(',')
st.sort()
for i in st:
    print(i)
print(len(st))
```
