Lets say our graph(Training Data) looks line this and we want to classify a given point, lets say X and and we want to classify this as either Triangle or Circle, in that case we find the nearest neighbors.  
What points are closest to testing data. What training data is closest to testing data.  
Lets say K = 1, what we will do is we will find the nearest datapoint and we will just give it a class same as that neighbour.  
So we will say that this is a Triangle.  
So if K = 3, we can see that 3 closest are one Triangle and Two Circles. And in this case, we end up saying it is a Circle.  
SO WHAT KNN SAYS IS LETS FIND THE FEW(CLOSEST) NEIGHBOUR AND ONCE WE FIND IT, WE CAN DECIDE THE CLASS ON THE BASIS OF A VOTEOUT (MAJORITY VOTE).  
WHY DO WE TAKE K AND NOT 1 NEAREST NEIGHBOUR?  
COZ 1 NEAREST NEIGHBOUR IS VERY PRONE TO OUTLIERS AND OVERFITTING.

But if we choose k as 3, then it will be classified correctly as cross. So we need to decidde this optimal K and we will do it.
The idea is to choose the k in a smart way rather than just choosing 1.  
Generally, we need to take k as odd coz lets say if we take k as 2, we might get stuck by classifying 1 from dot and 1 from cross.  
Few things to take care is what is the metric how do you define the neighbours, how close I am to somebody.  
We need to define distance.  
Distance can be Euclidean or Manhattan. Manhattan distance is defined as $\sum_{i=1}^i|x_{1}^i - x_{2}^i|$.(Sum of featurewise subtraction and modulo).  
Euclidean distance can be defined as $\sqrt(\sum_{i = 1}^n(x_{1}^i - x_{2}^i)^2 )$.  
So we will have to choose the k, we will have to choose the metric.  
And weightage of the vote can be defined as $\frac{1}{distance}$. This means vote is inversely proportional to distance.  
We can either go with uniform voting or weighted voting.  

KNN does more computation on test time than on train time?  
Solution Description  
There is no explicit training phase in KNN . It just takes data as input in its training phase . All the actual work , that is calculation of distances,comparisons and taking out nearest k neighbors is done at testing phase when test data is available
