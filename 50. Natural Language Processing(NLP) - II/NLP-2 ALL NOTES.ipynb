#### *Now we want to clean this files. By cleaning, we means we need to remove stopowrds and do lemmatization.*

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import string
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))
punctuation = list(string.punctuation)
stops.update(punctuation)
```

*Created stopwords and lemmatizer above*

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

**Explaination of this Function**  
This function will get the words.  
We need to clean these words.  
For that we need to create new array.
1. For each word in words, we need to check if w.lower() is in stops or not? if not,
2. If not, we will find the pos_tag. Now we need to pass the word w as it is without doing w.lower().  
   Because if we lower it, it might change the meaning of the word.  
3. Once we have the part of speech, we want to lemmatize it. If we know, pos cannot be passed directly, because  
   it requires to be different. So we use the get_simple_pos() function above to get the pos and then within that 
   we extract the pos for which the problem is solved in above cell with pos_tag.  
4. Then we will have the clean word which we get from lemmatizer and then we append it in the array and if we want, we can lower 
   it and append it.

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

Now in this we go through each document, category in documents we have, and then we want to add new tuple in the documents array where the first word in the document is the cleaned words from the function and second is the category.

```python
## To find the frequency distribution, 
freq = nltk.FreqDist(all_words)  ## this is a freq distribution object
```

*Now we have our features what do we want to do now is for each document, we want to create dictionary where this dictionary will have for eg: lets say for word film, {'film' : True or not} It means whether it comes in the document or not.If it comes the value of key is True else False.*  
We will ceate function which takes the words, and we will create dictionary for each document, then we will go through each words in the top 3000 features we created, we need to check if this word comes in the particular document, if this word w is the part of the array words(which are mentioned in the function), then we need to put a true else we need to put a false.  
We need to create a set of words.(which are of particular document we pass).


```python
def get_feature_dict(words):
    current_features =  {}
    words_set = set(words)
    for w in features:  
        current_features[w] = w in words_set   ## this will return True if w is in words_set else False
    return current_features
```

# 4. N-gram(Lecture)

**Different options available for CountVectorizer.**  
Only thing we saw about CountVectorizer is max_features. That is no of features that we can use.  
Another option that we can use is ngram.  
Another option is analyzer and by default, analyzer is word. If we want to break around character, we can do that.  
Lets focus on ngram.  
This is a very good technique that can lead to very god results in few cases.  
What exactly ngram is? By default what we are doing is 1-gram.  
By 2-gram we mean that if the text is 'the sky is blue', 2-gram means we will have 'the sky', 'sky is', 'is blue' as possibe words as well, the combination as possible feature as well.  
So if we were getting 'not good', then this will become one of the features as well and hopefully, this will relate to negative reviews. Now 'the sky' might not be very frequent, but 'not good' or 'very bad' might be very common combination of word.  
While we were selecting the top features, along with each individual word, you will be considering these bi-grams as well.  
So in n-gram  we can  give an option (1, 3) which means we want 1-grams, 2-grams and 3-grams as well.  
(2,3) means we need only 2-grams and 3-grams.  

*Number of N-grams for a sentence with X words will be X - N + 1*   
*Use  of N-Grams in your feature space may not necessarily yield any significant improvement.*


```python
count_vectorizer1 = CountVectorizer(max_features = 2000, ngram_range = (2,3))
x_train_features = count_vectorizer1.fit_transform(x_train)
x_train_features.todense()
```

# 5. TF-IDF(Lecture)


TF-IDF.  
TF means Term Frequency.  
IDF means Inverse Document Frequency.  
We only need to focus on Document Frequency and we will do $\frac{1}{Document Frequency}$ to get the IDF.  
  
So for a word w, its document frequency df(w) is defined as Number of Documents that contains w. The point is if a word is present in lots of document, its not that important.  
We want to choose a word which have higher frequency but we want to choose those words which have low document frequencies.  
They should be present at a lot of places, but they should not be present everywhere.  
Ideally a word w which comes in few of the documents but many times in those documents is far bettter than a word $w_2$ which comes in every document once. Because this word $w_2$ doesnt dive us any information whereas this w gives us a lot of information.  
So a lot of times we will be looking at number which is $tf * idf$.  
When we use countvectorizer, there are couple of very interesting argument that we can give.  
There is an argument which is max _df and min_df.  
This basically means when we are building our vocabulary we can say hey I dont want to take any document, I dont want to take any word in my vocabulary which has more than 0.8 df.  
That means if a word is present in more than 80% document, We are not interested in that word.  
Or if a document is present in less than 10% document, we are not interested in that word.  
We can choose max_df and min_df to fingure out what kind of word we dont want to keep.  
This will help us choose the vocabulary.  
Even if we dont remove stopwords and use max_df, it will kind of remove stopwords.  
One good thing about this is at even if there is a word which is not very common in english.  
For eg, lets say email is not a stopword and it is not, but lets say we have a dataset, which comprises of a lot of email, now here we feel that a lot more document will have word email in them.  
So the document frequency of the email is very high for this set of documents.  
Or if we are looking at the documents like emails sent accross a university, it might have word like class, grade which might be very common and might not give us any information, so using this max_df and min_df we might be able to get rid of few words from our vocabulary which might be behaving like a stopwords in the context of the current document.  
Similarly we might be able to get rid of some words which are just present in like 5-7 documents but like full completely repeated many time in those 5-7 documents but they wont give us much information because they are present in very few documents.  
So we can use max_df, min_df that present with us n countVectorizer to basically avoid both category of words.  
That being said there is another level of using tf-idf. min_df and max_df is used during the time when we built the vocabulary.  
Once the vocabulary is ready, we have our feature $f_1$ to $f_n$, sometimes what will work better for you is, right now we are using tf, for $d_1$, what is the $tf(w_1, d_1)$ ..... $tf(w_{n}, d_1)$  which means how many times $w_1$ comes in $d_1$ till how many times $w_n$ comes in $d_1$.  
So in $d_{10}$, the tf will be $tf(w_1, d_{10})$ .... $tf(w_n, d_{10})$ which means how many times $w_1$ comes in $d_10$ till how many times $w_n$ comes in $d_10$.  
Sometimes it will be better if we use $tf(w_1, d_n) * idf $  instead of $tf(w_1, d_n)$.  
    The idea again here is that just because somehow email is still part of the vocabulary, and email got lets say $d_1$ contains email 3 times, it will make sense that we will multiply with idf where idf is very low because document frequency will be very high. If email is present everywhere, the document frequency will be very high and idf will be very low and weightage of this feature will go down.  
So we can use tf-idf instead of using noraml countvectorizer. For that, we have tfidfvectorizer if we want to use it.  
So if we want to create the feature in $tf(w_1, d_n) * idf $ format, we will use tfidfvectorizer.  
But also we need to experiment it with max_df min_df we have in countvectorizer.  
max_df will help get rid of words which acts like stopwords.
