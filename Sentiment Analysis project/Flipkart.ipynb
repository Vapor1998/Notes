```python
def link_collections():
    
    driver.execute_script('window.scrollBy(0, document.body.scrollHeight);')
    time.sleep(2)
    review = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME,'_1fQZEK')))
    title = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME,'_4rR01T')))

    titles = [] 
    d = {}
    reviews = []

    for i in review:
        if '(' in i.text:
            reviews.append(i)

    for i, j in zip(title, reviews):
        x = search.lower()
        if x in i.text.lower() and '(Renewed)' not in i.text:
            d[j] = i.text
            titles.append(i.text)
            
    time.sleep(1)
    return d
```

```python
def link_collection():
    
    driver.execute_script('window.scrollBy(0, document.body.scrollHeight);')
    time.sleep(2)
    review = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME,'_1fQZEK')))
    title = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME,'_4rR01T')))

    titles = [] 
    d = {}
    reviews = []
    
    try:
        for i in review:
            if '(' in i.text:
                reviews.append(i)
    except StaleElementReferenceException:
        driver.refresh()
        time.sleep(2)
        link_collection()
    try:
        for i, j in zip(title, reviews):
            #x = search.lower()
            if '(Renewed)' not in i.text:
                d[j] = i.text
                titles.append(i.text)
    except StaleElementReferenceException:
        driver.refresh()
        time.sleep(2)
        link_collection()
        
    if len(d) == 0:
        driver.refresh()
        time.sleep(2)
        link_collection()
        
        
            
    time.sleep(1)
    return d
```

```python

def scraping(d, limit):
    j = 1
    prod_review= {}
    link = {}
    overall_rating = {}
    price = {}
    ratings = {}
    
    for i in d:
        driver.execute_script('window.scrollBy(0, document.body.scrollHeight);')
        driver.switch_to.window(driver.window_handles[0])
        i.click()
        time.sleep(2)
        
        #driver.switch_to.window(driver.window_handles[j])
        c = driver.window_handles[1]
        driver.switch_to.window(c)
        y = d[i]
        
        # scraping link
        url = driver.current_url
        link[y] = url
        
        
        # scraping stars
        e = driver.page_source
        x= BeautifulSoup(e, 'html.parser')
        h = x.find_all('div', {'class' : '_1uJVNT'})
        stars = {}
        count = 5
        for i in h:
            stars["{} Stars".format(count)] = i.text
            count-=1
            
        ratings[y] = stars
        
        
        #scraping price   <div class="_30jeq3 _16Jk6d">â‚¹69,999</div>
        e = driver.page_source
        x= BeautifulSoup(e, 'html.parser')
        h = x.find_all('div', {'class' : '_30jeq3 _16Jk6d'})
        p = []
        for i in h:
            p.append(i.text)
        price[y] = p
        
        
        # average ratings
        e = driver.page_source
        x= BeautifulSoup(e, 'html.parser')
        h = x.find_all('div', {'class' : '_2d4LTz'})
        r = []
        for i in h:
            r.append(i.text)
        overall_rating[y] = r
        
        
        driver.execute_script('window.scrollBy(0, document.body.scrollHeight);')
        time.sleep(7)
        all_comments_link = WebDriverWait(driver, 6).until(EC.presence_of_element_located((By.CLASS_NAME, '_3UAT2v._16PBlm'))) 
        driver.execute_script("arguments[0].click();", all_comments_link)
        
        # read more 
        # work in progress
#         flag = True
#         while flag:
#             try:
#                 read_more = WebDriverWait(driver, 5).until(EC.presence_of_all_elements_located((By.CLASS_NAME, '_1BWGvX')))
#                 if len(read_more) == 0:
#                     flag = False
#                     break
#                 for i in read_more:
#                     i.click()
#                     time.sleep(1)
#                 flag = False
#             except TimeoutException:
#                 driver.refresh()
#                 pass
        
        
        #time.sleep(3)
        
        
        
        
        # scraping review
        e = driver.page_source
        x= BeautifulSoup(e, 'html.parser')

        h = x.find_all('div', {'class' : 't-ZTKy'})
        reviews = set()
        
        for k in h:
            reviews.add(k.text)
            if len(reviews) == limit:
                break
        

        if len(reviews) != limit or len(reviews) < limit:
                try:
                    while len(reviews) != limit:
                        review1 = reviews
                        next_page = WebDriverWait(driver, 5).until(EC.presence_of_all_elements_located((By.CLASS_NAME, '_1LKTO3')))
                        if len(next_page) == 1:
                            next_page[0].click()
                        elif len(next_page) == 2:
                            next_page[1].click()
                        time.sleep(4)
                        driver.execute_script('window.scrollBy(0, document.body.scrollHeight);')
                        e = driver.page_source
                        x= BeautifulSoup(e, 'html.parser')
                        t = x.find_all('div', {'class' : 't-ZTKy'})
                        for l in t:
                            reviews.add(l.text.strip())
                            if len(reviews) == limit:
                                break
                        prod_review[y] = reviews

                except TimeoutException:
                    pass
        else:
            prod_review[y] = reviews
            
        #prod_review[y] = reviews
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
            
        
        print(j)
        j+=1
        
    return prod_review, link, overall_rating, price, ratings
```
