```python
import matplotlib.pyplot as plt
import numpy as np

year = [2012, 2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]
```

```python
## Bubble graph is basically built by using scatter() where size of the bubble varies according to the
## information of graph.
## Size of the bubble also represents information of graph.
## So to make it, we actually use s = arrayname  in scatter() to give size to
## particular bubble as per the information


plt.scatter(year,salary, s = population)
plt.title("Bubble Graph")
plt.show()
```

```python
## To maintain the opacity of the color, we pass alpha = value as argument in scatter().

plt.scatter(year,salary, s = population, c = 'r', alpha = 0.5)
plt.title("Bubble Chart")
plt.show()

## now we can see the the color is dull or less opaque.
```

```python
## To make different color of different values, we can pass array in c = value as value 
## in scatter() instead of color value.
## Make sure the value passed in array is as same as no of bubbles of graph




a = np.arange(len(year))

plt.scatter(year,salary, s = population, c = a)
plt.title("Bubble Graph")
plt.show()

## To pass different color, we used arange() where we passed length of year array 
## and we passed arange() array in c.
## If we have same value in c array, it will give same color.
```

```python
## Edge color
## Edge color will decide edge color of the bubble in graph.

plt.scatter(year,salary, s = population, c = a, edgecolor = "black")
plt.title("Bubble Graph")
plt.show()

## here we can see the edgecolor of the bubbles are black
```

```python
## Change the marker
## We can also change the shape of bubble by passing marker = 'value' as argument in 
## the scatter()

plt.scatter(year,salary, s = population, c = a,alpha = 0.5, marker = '^')

plt.title("Bubble Graph")
plt.grid()
plt.show()

## As we can see, shape of the bubble has changed.
```

```python
## Add the text.
## To add the text, we use text()
## It takes argument text(x, y, 'text') where x & y are the location where we want to put the text.

plt.scatter(year,salary, s = population, c = a,alpha = 0.5, marker = '^')
plt.title("Bubble Graph")
plt.text(year[0], salary[0], "100")
plt.grid()
plt.show()
```

```python
## As we can see, value is overlapping to the marker, so we can fix that by adding
## extra values in the x and y.

plt.scatter(year,salary, s = population, c = a,alpha = 0.5, marker = '^')
plt.text(year[0] + 0.1, salary[0]+ 0.1, '100')    ## here, we increased the x and y axis value by 0.1
plt.title("Bubble Graph")
plt.show()

## As we can see, it is fixed.
```

```python
## We can also pass array value in text

plt.scatter(year,salary, s = population, c = a,alpha = 0.5, marker = '^')
plt.text(year[0] + 0.1, salary[0]+ 0.1, population[0])    ## here, we increased the x and y axis value by 0.1
plt.text(year[4] + 0.1, salary[4] + 0.1, population[4])
plt.title("Bubble Graph")
plt.show()
```
