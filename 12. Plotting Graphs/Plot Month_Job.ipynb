```python
## Few concepts here

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from string import digits

amazon = pd.read_csv("amazon_jobs_dataset.csv")

df = amazon.copy()

date = df["Posting_date"]
date = date.str.split(' ', expand = True)  ## This split() with expand = True expands the string into 
                                           ## seprate columns.
month = date[0]
freq = month.value_counts()
months = freq.index
jobs = freq.values

plt.bar(months, jobs)
plt.xticks(rotation = 90)
plt.show()

for i in range(len(months)):
    print(months[i], jobs[i])

```

```python
# Open and read data file as specified in the question
# Print the required output in given format
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from string import digits

amazon = pd.read_csv("amazon_jobs_dataset.csv")

df = amazon.copy()

month = df["Posting_date"].str.replace('\d+','')   ## To remove all the numbers from string.Useful concept.
months = month.str.replace(',','')
months = months.str.rstrip()    ## Use to remove whitespaces in right side of string 
months.value_counts()

month_name = months.index
jobs = months.values

plt.bar(jobs, month_name)
plt.show()
```
