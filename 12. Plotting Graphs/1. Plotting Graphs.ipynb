```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]
y = [2,4,6]
## above line falls on the line y = 2x

plt.scatter(x,y)    ## This function scatters the points over the graph.But it will not show graph.
                    ## Actually, this function will create a point on graph.

plt.show()     ## This will show the graph.
```

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)    ## plot() will show a line on the graph.
                 ## Actually it will connect the points on the graph.
    
plt.show()    ## This will show the graph.
```

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)    
plt.scatter(x,y)   

## Above plot() will plot a point on the graph while the scatter() will draw the line on the graph to
## connect the points.

plt.show()


```

```python
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x,y)
plt.plot(x2,y2)

## We have plotted four variables i.e. two lines on the graph.

plt.show()
```

```python
## To change the color of graph

plt.plot(x,y, "r--")   ## The first character of the string will decide the color of the graph,
                      ## the second character of the string will decide how you want your graph,
                     ## whether line or dots., In this case, it is dashed lines.
                    ## Read more for documentation.
        
plt.show()
```

```python
## plot function x^3   (xcube)

x = np.array([1,2,3,4])
y = x**3

plt.plot(x,y,'b')
plt.show()

## It looks like line, but to make it look like curve, we use np.arange()
```

```python
x = np.arange(0,5,0.2)
y = x**3

plt.plot(x,y, 'b-')
plt.show()

## Now it looks like a curve
```

```python
## Case 
## If we give only one array, a , it will treat array as y, 
## while it will take x as range(0,n,1) where n is number of elements in a.

a = [3,4,5]
plt.plot(a)
plt.show()


```
