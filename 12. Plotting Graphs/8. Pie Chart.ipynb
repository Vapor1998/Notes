```python
import matplotlib.pyplot as plt
import numpy as np


sizes = [3,4,6,2]
plt.pie(sizes)     ## This will create a pie graph.
plt.show()

## In our case, diameter and radius of pie graph is equal.
```

```python
## If the pie chart diameter and radius is not equal, we will pass axis('equal') before show().

sizes = [3,4,6,2]
plt.pie(sizes)     ## This will create a pie graph.
plt.axis('equal')
plt.show()
```

```python
## To pass different color.
## To pass different color, we will pass an array with different color names to color = arrayname in the pie() as argument

sizes = [3,4,6,2]
color = ['blue', 'purple', 'red', 'pink']
plt.pie(sizes, colors = color)     ## This will create a pie graph.
plt.axis('equal')
plt.show()
```

```python
## To add labels.
## To pass label, we will pass an array with different label names to labels = arrayname in the pie() as argument


sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
plt.pie(sizes, colors = color, labels = label)     ## This will create a pie graph.
plt.axis('equal')
plt.show()
```

```python
## Title
## To add title in pie chart, we use title("Titlename") before show().
sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
plt.pie(sizes, colors = color, labels = label)     ## This will create a pie graph.
plt.title("Split among classes")
plt.axis('equal')
plt.show()
```

```python
## To add what is the percentage split among classes,
## we use autopct as argument in pie().
## Inside it, we can pass arg as '%.1f' which means it will show how much digit to show after decimal 
## point in percentage

sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
plt.pie(sizes, colors = color, labels = label, autopct = '%.2f%%')     ## This will create a pie graph.
plt.axis('equal')
plt.title("Split among classes")
plt.show()

## Two percentage in the string '%.2f%%' will add percentage sign at the end of autopct.
```

```python
## Explode()
## explode() will actually cut out the portion of graph by particular distance passed to it.
## To use it we will pass array where we say how much distance we want to cut a particular portion
## and then pass it to the explode = arrayname as the argument in the pie().

sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
explodes = [0.1,0,0,0]
plt.pie(sizes, colors = color, labels = label, autopct = '%.2f%%', explode = explodes)     ## This will create a pie graph.
plt.axis('equal')
plt.title("Split among classes")
plt.show()

## As we can see, A has been cut out by 0.1 distance from pie chart.
```

```python
## If we see our graph, it is anticlockwise.
## To make it clockwise, we pass argument counterclock = False in pie().
## counterclock is by default True.

sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
explodes = [0.1,0,0,0]
plt.pie(sizes, colors = color, labels = label, autopct = '%.2f%%', explode = explodes, counterclock = False)     
plt.axis('equal')
plt.title("Split among classes")
plt.show()
```

```python
## The startangle of the A is 0 degree.
## To change the degree, we use startangle = angle as argument in pie() and define the angle in the argument from where
## we want to start.

sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
explodes = [0.1,0,0,0]
plt.pie(sizes, colors = color, labels = label, autopct = '%.2f%%', explode = explodes, counterclock = False, startangle = 100)     
plt.axis('equal')
plt.title("Split among classes")
plt.show()

## As we can see, A is starting from angle 100.
```
