```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from string import digits

amazon = pd.read_csv("amazon_jobs_dataset.csv")

df = amazon.copy()

country = df['location'].str.split(' ', expand = True)
ind = country[country[0] == 'IN,']
cities = ind[2].value_counts()
city = cities.index
jobs = cities.values

plt.pie(jobs, labels = city, autopct = '%.2f%%')
plt.show()

for i in range(len(jobs)):
    percent = jobs[i] /sum(jobs) *100
    print(city[i], format(percent, '.2f'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
amazon = pd.read_csv("amazon_jobs_dataset.csv", encoding = 'latin1')
df = amazon.copy()

df=df['location'].str.split(",",expand=True)
df.columns=["Country","State","City"]
df=df[df["Country"]=="IN"]
df.drop(["State","Country"],axis=1)

count=df['City'].value_counts()

city=count.index
value=count.values


plt.pie(value,labels=city,autopct="%.2f%%")
plt.show()

sum=np.sum(value)
value_float=[]
for i in range (len(value)):
    a=float((value[i])*100/sum)
    value_float.append(a)
    
for i in range(len(value_float)):
    print (city[i].strip(),format(value_float[i], '.2f'))

    
```

```python
# Open and read data file as specified in the question
# Print the required output in given format
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


def cal(np_freq):
    percent_list = []
    np_sum = np.sum(np_freq)
    for i in np_freq:
        percent = (i/np_sum)
        percent = round((percent * 100),3)
        percent_list.append(percent)
    np_percent = np.array(percent_list)
    return np_percent
    
    
file_obj = open('amazon_jobs_dataset.csv',encoding='utf8')
data = pd.read_csv(file_obj,skipinitialspace=True)
countries = data["location"].str.split(',',expand=True)
country = countries[0]
cities = countries[2]
country_india = np.where(country == "IN")[0]
cities_india = cities[country_india]
freq = cities_india.value_counts()
label = freq.index
freq.sort_values(ascending = False,inplace=True)
np_freq = np.array(freq)
percent_list = cal(np_freq)
plt.pie(freq,labels = label,autopct = "%.2f")
plt.axis("equal")
#plt.title("Indian Cities vs Percentage Jobs",fontsize=12)
plt.show()
india_cities = freq.index
for i in range(len(freq)):
    print(india_cities[i].strip(),format(percent_list[i],'.2f'))
```
