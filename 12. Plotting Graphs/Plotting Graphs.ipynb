# 4. Customizing Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,5,0.2)
y = x**3

plt.plot(x,y, color = 'blue')    ## One more way to pass color
plt.show()
```

```python
## Marker 
## To add markers in the graph, we need to pass arg 'marker' in plot() function.
## Markers can be of type +, o, p, etc.
## These are the markers alongside with the line connecting with it.

x = np.arange(0,5,0.2)
y = x**3

plt.plot(x,y, color = 'blue', marker = 'o')    
plt.show()
```

```python
## LineWidth
## This will allow us to give linewidth to the line in the graph
## To use it, we add 'linewidth' argument in plot() function.\



x = np.arange(0,5,0.2)
y = x**3

plt.plot(x,y, color = 'blue', linewidth = 5)    
plt.show()

## You can see, line is thicker than the above graph.
```

```python
## YLabel & XLabel & Title
## This will add the label(Legend) accross the x axis and y axis.
## Title will add title to the graph.
## To add it, we use ylabel('labelname'), xlabel('labelname'), title('titlename') before show().
 


x = np.arange(0,5,0.2)
y = x**3

plt.plot(x,y, color = 'blue', marker = 'o')    
plt.ylabel('X^3')
plt.xlabel('X')
plt.title("Matplotlib demo")
plt.show()
```

```python
## Plotting two lines in graph
x = np.arange(0,5,0.2)
y = x**3
y2 = x **2

plt.plot(x,y, color = 'blue')
plt.plot(x,y2, color = 'black')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
## By looking at the graph, if we have two lines, we can identify which graph is which
## So to do that, we seprately pass label as argument in the plot() for the particular line
## and then pass label inside it and then pass legend() before pass.


x = np.arange(0,5,0.2)
y = x**3
y2= x**2

plt.plot(x,y, color = 'blue', label = 'x^3')
plt.plot(x,y2, color ='black', label = 'x^2')
plt.ylabel('y')
plt.xlabel('x')
plt.legend()
plt.show()

## Now as we can see, we can differentiate between two graph as we can
## see legend.
```

```python
## axis()
## If we want to see more axis of graph, we use axis()
## axis() takes four arguments inside list, axis([startX, endX, startY, endY])


x = np.arange(0,5,0.2)
y = x**3
y2= x**2

plt.plot(x,y, color = 'blue', label = 'x^3')
plt.plot(x,y2, color ='black', label = 'x^2')
plt.axis([0,10,0,200])
plt.ylabel('y')
plt.xlabel('x')
plt.legend()
plt.show()

## As we can see, we can see more axis of graph.
```

```python
## grid()
## This will create grid in graph.

x = np.arange(0,5,0.2)
y = x**3
y2= x**2

plt.plot(x,y, color = 'blue', label = 'x^3')
plt.plot(x,y2, color ='black', label = 'x^2')
plt.ylabel('y')
plt.xlabel('x')
plt.grid()
plt.legend()
plt.show()
```

```python
## To add random text at some place.
## To add text, we use tex(), it takes three argument
## text(x, y, "text")  where text is what we want to add at x, y.

x = np.arange(0,5,0.2)
y = x**3
y2= x**2

plt.plot(x,y, color = 'blue', label = 'x^3')
plt.plot(x,y2, color ='black', label = 'x^2')
plt.ylabel('y')
plt.xlabel('x')
plt.text(2, 80,'Hii')
plt.grid()
plt.legend()
plt.show()
```

# 6. Bubble Chart

```python
import matplotlib.pyplot as plt
import numpy as np

year = [2012, 2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]
```

```python
## Bubble graph is basically built by using scatter() where size of the bubble varies according to the
## information of graph.
## Size of the bubble also represents information of graph.
## So to make it, we actually use s = arrayname  in scatter() to give size to
## particular bubble as per the information


plt.scatter(year,salary, s = population)
plt.title("Bubble Graph")
plt.show()
```

```python
## To maintain the opacity of the color, we pass alpha = value as argument in scatter().

plt.scatter(year,salary, s = population, c = 'r', alpha = 0.5)
plt.title("Bubble Chart")
plt.show()

## now we can see the the color is dull or less opaque.
```

```python
## To make different color of different values, we can pass array in c = value as value 
## in scatter() instead of color value.
## Make sure the value passed in array is as same as no of bubbles of graph




a = np.arange(len(year))

plt.scatter(year,salary, s = population, c = a)
plt.title("Bubble Graph")
plt.show()

## To pass different color, we used arange() where we passed length of year array 
## and we passed arange() array in c.
## If we have same value in c array, it will give same color.
```

```python
## Edge color
## Edge color will decide edge color of the bubble in graph.

plt.scatter(year,salary, s = population, c = a, edgecolor = "black")
plt.title("Bubble Graph")
plt.show()

## here we can see the edgecolor of the bubbles are black
```

```python
## Change the marker
## We can also change the shape of bubble by passing marker = 'value' as argument in 
## the scatter()

plt.scatter(year,salary, s = population, c = a,alpha = 0.5, marker = '^')

plt.title("Bubble Graph")
plt.grid()
plt.show()

## As we can see, shape of the bubble has changed.
```

```python
## Add the text.
## To add the text, we use text()
## It takes argument text(x, y, 'text') where x & y are the location where we want to put the text.

plt.scatter(year,salary, s = population, c = a,alpha = 0.5, marker = '^')
plt.title("Bubble Graph")
plt.text(year[0], salary[0], "100")
plt.grid()
plt.show()
```

```python
## As we can see, value is overlapping to the marker, so we can fix that by adding
## extra values in the x and y.

plt.scatter(year,salary, s = population, c = a,alpha = 0.5, marker = '^')
plt.text(year[0] + 0.1, salary[0]+ 0.1, '100')    ## here, we increased the x and y axis value by 0.1
plt.title("Bubble Graph")
plt.show()

## As we can see, it is fixed.
```

```python
## We can also pass array value in text

plt.scatter(year,salary, s = population, c = a,alpha = 0.5, marker = '^')
plt.text(year[0] + 0.1, salary[0]+ 0.1, population[0])    ## here, we increased the x and y axis value by 0.1
plt.text(year[4] + 0.1, salary[4] + 0.1, population[4])
plt.title("Bubble Graph")
plt.show()
```

# 8. Pie chart

```python
import matplotlib.pyplot as plt
import numpy as np


sizes = [3,4,6,2]
plt.pie(sizes)     ## This will create a pie graph.
plt.show()

## In our case, diameter and radius of pie graph is equal.
```

```python
## If the pie chart diameter and radius is not equal, we will pass axis('equal') before show().

sizes = [3,4,6,2]
plt.pie(sizes)     ## This will create a pie graph.
plt.axis('equal')
plt.show()
```

```python
## To pass different color.
## To pass different color, we will pass an array with different color names to color = arrayname in the pie() as argument

sizes = [3,4,6,2]
color = ['blue', 'purple', 'red', 'pink']
plt.pie(sizes, colors = color)     ## This will create a pie graph.
plt.axis('equal')
plt.show()
```

```python
## To add labels.
## To pass label, we will pass an array with different label names to labels = arrayname in the pie() as argument


sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
plt.pie(sizes, colors = color, labels = label)     ## This will create a pie graph.
plt.axis('equal')
plt.show()
```

```python
## Title
## To add title in pie chart, we use title("Titlename") before show().
sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
plt.pie(sizes, colors = color, labels = label)     ## This will create a pie graph.
plt.title("Split among classes")
plt.axis('equal')
plt.show()
```

```python
## To add what is the percentage split among classes,
## we use autopct as argument in pie().
## Inside it, we can pass arg as '%.1f' which means it will show how much digit to show after decimal 
## point in percentage

sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
plt.pie(sizes, colors = color, labels = label, autopct = '%.2f%%')     ## This will create a pie graph.
plt.axis('equal')
plt.title("Split among classes")
plt.show()

## Two percentage in the string '%.2f%%' will add percentage sign at the end of autopct.
```

```python
## Explode()
## explode() will actually cut out the portion of graph by particular distance passed to it.
## To use it we will pass array where we say how much distance we want to cut a particular portion
## and then pass it to the explode = arrayname as the argument in the pie().

sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
explodes = [0.1,0,0,0]
plt.pie(sizes, colors = color, labels = label, autopct = '%.2f%%', explode = explodes)     ## This will create a pie graph.
plt.axis('equal')
plt.title("Split among classes")
plt.show()

## As we can see, A has been cut out by 0.1 distance from pie chart.
```

```python
## If we see our graph, it is anticlockwise.
## To make it clockwise, we pass argument counterclock = False in pie().
## counterclock is by default True.

sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
explodes = [0.1,0,0,0]
plt.pie(sizes, colors = color, labels = label, autopct = '%.2f%%', explode = explodes, counterclock = False)     
plt.axis('equal')
plt.title("Split among classes")
plt.show()
```

```python
## The startangle of the A is 0 degree.
## To change the degree, we use startangle = angle as argument in pie() and define the angle in the argument from where
## we want to start.

sizes = [3,4,6,2]
label = ['A','B',"C",'D']
color = ['blue', 'purple', 'red', 'pink']
explodes = [0.1,0,0,0]
plt.pie(sizes, colors = color, labels = label, autopct = '%.2f%%', explode = explodes, counterclock = False, startangle = 100)     
plt.axis('equal')
plt.title("Split among classes")
plt.show()

## As we can see, A is starting from angle 100.
```

# 10. Histogram

## Histogram  
A histogram is a graph showing frequency distributions.  
  
It is a graph showing the number of observations within each given interval.  

```python
import matplotlib.pyplot as plt
import numpy as np

a = [1,2,3,4,1,3,4,2,1,1]
plt.hist(a)
plt.show()

## This histogram chart is showing frequency of each number on which number is coming how many times.
## For eg, 1 is coming 4 times, 2 is coming 2 times etc
```

```python
b = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.hist(b)
plt.show()

## Now if we see here, 1 is not showing 6 times, but 1 is showing 2 times, 2 is showing 1 time, and 3 is showing 
## 3 times.
## So what happened here is it has clubbed the number together to show the frequencies.
```

```python
## TO see more clear, we use edgecolor = 'black' which actually colors the edge of histogram
## This is passed as argument in hist()

b = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.hist(b, edgecolor = 'black')
plt.show()

## Seems like it has clubbed numbers inside one bin.
## To understand what numbers have been clubbed, we use xticks().
```

```python
## xticks() takes array as argument and is used before show().
## xticks is used to show numbers in histogram in x axis.

b = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.hist(b, edgecolor = 'black')
x = np.arange(23)
plt.xticks(x)
plt.show()

## Here, our histogram has numbers from 1 to 22 so we need numbers from 1 to 22.
## To do that, we use arange() and then we pass that arange to xticks().

## If we look carefully, bin1 is from 1 to 3.1 as it is slightly away from 3, and bin2 is from 3.1 to 5.2,
## and so on.
## Bydefault each histogram creates 10 towers i.e. 10 bins.
## Size of the bin is calculated by (maxNumber - minNumber)/no of bins.
## We can calculate it in our case.


```

```python
## We can also tell how much bins we want by passing bin = value as arg in hist().
## where value is no of bins we want.

b = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.hist(b,bins = len(b), edgecolor = 'black')
x = np.arange(23)
plt.xticks(x)
plt.show()

## as we can see, bins are located at [a,b].
## for eg, look at the first bin, it is located at [1,2]
## and we can see that all the bins are in the same format.
## To identify that, [a,b] means include a , exclude b.
## So to identify which number bin belongs to, we can say it that way.
## For e.g. bin1 is from[1,2] which means that bin is for number 1.
## Similarly, bin2 is from[2,3] which means that bin is for number 2.
## Except for last case, where we dont have number 21 in array, but it is showing number 21.
```

```python
## we can also define our own bins by adding bins = [ values ] in hist()
## We can also change the axis limit by using axis([xmin, xmax, ymin,ymax]).
plt.hist(b, bins = [1,3,6,10,15,20], edgecolor = 'black')
plt.axis([0,25,0,10])
plt.show()
```

# Plot Job_Java

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from string import digits

amazon = pd.read_csv("amazon_jobs_dataset.csv")

df = amazon.copy()

year = df['Posting_date'].str[-4:]
string  = 'Java'
string2 = 'java'

javaJobs = df[df["BASIC QUALIFICATIONS"].str.contains(string,string2, na = True)]     ## Concept for finding particular value in string
year = javaJobs['Posting_date'].str[-4:]
year = year.value_counts()

years = year.sort_index()

yearCount = years.index
jobs = years.values

plt.scatter(yearCount, jobs)
plt.show()

for i in range(len(jobs)):
    print(yearCount[i], jobs[i])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
amazon = pd.read_csv("amazon_jobs_dataset.csv")

df = amazon.copy()

years = df['Posting_date'].str[-4:]
counts = years.value_counts()
counts.sort_index(inplace = True)

year = counts.index
jobs = counts.values

plt.plot(year, jobs)
plt.show()

for i in range(len(year)):
    print(year[i], jobs[i])
```

```python
# Open and read data file as specified in the question
# Print the required output in given format
import csv
import matplotlib.pyplot as plt
import numpy as np
with open('amazon_jobs_dataset.csv', encoding ='latin1') as file_obj:
    file_data=csv.DictReader(file_obj)
    d={}
    for rows in file_data:
        year=rows['Posting_date'].split(',')[1].strip()
        if year in d:
            d[year]+=1
        else:
            d[year]=1
k=dict(sorted(d.items(),key=lambda item:item[0]))
years=[]
jobs=[]
for i in k:
    years.append(i)
    jobs.append(d[i])
plt.plot(years,jobs)
plt.show()
for i in years:
    print(i,d[i])
```

# Plot Month_Job

```python
## Few concepts here

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from string import digits

amazon = pd.read_csv("amazon_jobs_dataset.csv")

df = amazon.copy()

date = df["Posting_date"]
date = date.str.split(' ', expand = True)  ## This split() with expand = True expands the string into 
                                           ## seprate columns.

month = date[0]
freq = month.value_counts()
months = freq.index
jobs = freq.values

plt.bar(months, jobs)
plt.xticks(rotation = 90)
plt.show()

for i in range(len(months)):
    print(months[i], jobs[i])

```

```python
# Open and read data file as specified in the question
# Print the required output in given format
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from string import digits

amazon = pd.read_csv("amazon_jobs_dataset.csv")

df = amazon.copy()

month = df["Posting_date"].str.replace('\d+','')   ## To remove all the numbers from string.Useful concept.
months = month.str.replace(',','')
months = months.str.rstrip()    ## Use to remove whitespaces in right side of string 
months.value_counts()

month_name = months.index
jobs = months.values

plt.bar(jobs, month_name)
plt.show()
```
