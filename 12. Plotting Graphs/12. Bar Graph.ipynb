```python
import matplotlib.pyplot as plt
import numpy as np

year = [2012, 2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

plt.bar(year, population)
plt.show()
```

```python
## deciding width of bargraph
## to decide the width of the graph, we pass width = value as arg in bar().


plt.bar(year, population, width = 0.1)  ## default is 0.8
plt.show()
```

```python
year1 = ["year - 2012","year - 2013","year - 2014","year - 2015","year - 2016","year - 2017"]

plt.bar(year1, population)  
plt.show()

## if we look, x axis is not perfect, to make it, we can rotate the particular value.
```

```python
## To rotate a particular value, we use xticks(), inside it
## we pass rotation = degrees as arg to rotate value

year1 = ["year - 2012","year - 2013","year - 2014","year - 2015","year - 2016","year - 2017"]

plt.bar(year1, population) 
plt.xticks(rotation = 90)
plt.show()

```
