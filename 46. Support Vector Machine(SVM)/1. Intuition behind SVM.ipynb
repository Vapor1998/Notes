### SVM
It is a classification Algorithm.  
Its among the most powerful Algorithms we can use.  
What is SVM?  
Lets start with Logistic Regression. In case of Logistic Regression, our hypothesis function is (taking m as $\theta$)
$$h_{\theta}(x) = \frac{1}{1+e^{-\theta^Tx}}$$
When we draw this we draw this wrt $\theta^Tx$
When we plot this, we realise when $\theta^Tx$ = 0, function hits the 0.5 on Y-Axis and graph looks like below.  
So it goes towards 1 never reaches 1, it goes towards 0 never reaches 0.  



So if we have to make a choice between class 0 and class 1, what we will say is if our hypothesis $h_{\theta}(x)$ >= 0.5, We predict class 1, if $h_{\theta}(x)$ < 0.5, we predict class 0, which effectively means we are going to say if  => if $\theta^Tx$ >= 0, then our hypothesis is $h_{\theta}(x)$ >= 0.5 which is evident from the graph, this means class 1 and if $\theta^Tx$ < 0, this means class 0. Thats how we defined Logistic Regression.  
What SVM says is lets not just settle on 0, we should try to increase this margin a lot more.  
We should try incase y = 1 for our training data, we should try that $\theta^Tx$ >= some other value and not just 0.  
Lets try to avoid the particular area near graph and lets go beyond, So we generally choose the value to be 1.  
This basically means we want the predictions we want $\theta^Tx$ value to be > 1 and not just >= 0.  
We will still predict 1 in case it is 0.5 but the idea is the cost function somehow makes sure that we are keeping a large margin and we are lotmore confident when we say y = 1.  
And if y = 0, that case we want that $\theta^Tx$ <= -1.  
SO we want this to be huge, we want the gap to be big, we dont want our data to come somewhere in the middle, we want this to be more than this.  
And the only idea is if we are keeping the large margin, we are lot more confident.  
if the hypothesis says 0.5001, we are not that confident whether this is going to be class 0 or class 1.  
The idea is to make $\theta^Tx$ big enough such that $h_{\theta}(x) = \frac{1}{1+e^{-\theta^Tx}}$ either tends towards 1 or 0 so that we are lot more confident and we want make that integrate that into our cost function as well. Thats the basic Intitution of SVM.  
Lets say we did all of these and we still get $\theta^Tx$ to be 0.6, that does not mean that that we are going to predict some other class, this still means class 1. The whole point is we will change the cost function in a way that this is highly motivated that we will end up with  $\theta^Tx$ <= -1 margin.

## Large Margin Intuition.  
In logistic regression, we take the output of the linear function and squash the value within the range of [0,1] using the sigmoid function. If the squashed value is greater than a threshold value(0.5) we assign it a label 1, else we assign it a label 0. In SVM, we take the output of the linear function and if that output is greater than 1, we identify it with one class and if the output is -1, we identify is with another class. Since the threshold values are changed to 1 and -1 in SVM, we obtain this reinforcement range of values([-1,1]) which acts as margin.  
  
**We take values to be [-1, 1] because we want marginal distance to be greater. Thats the intuition.**
