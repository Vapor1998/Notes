## Gaussian Kernel  
It is defined as $$sim(x, l_{i}) = e^-({\frac{||x - l_{i}|| ^ 2}{2\sigma^2})}$$.  
Lets say we have a $l_{i}$, if the x is very close, the similarityy will be higher.  
And if x goes away, the similarity will start reducing.  
And because its exponential, it looks like  

where middle line is the $l_{i}$ which is the maximum value and it is 1 and the spread is determined by the $\sigma$ as stated in the graph.  

Now lets talk about $\sigma^2$.  
Sometimes $\frac{1}{2\sigma^2}$ is used as $\lambda$.  
So the equation looks like $$sim(x, l_{i}) = e^-(\lambda ||x - l_{i}|| ^ 2)$$.  
Lets say $\sigma^2$ is 1 and is representing above graph.  
Lets say the $\sigma^2$ is 10.  
So if we look at $$sim(x, l_{i}) = e^-({\frac{||x - l_{i}|| ^ 2}{2\sigma^2})}$$, where $\sigma^2$ is 10, it becomes $e^-({\frac{||x - l_{i}|| ^ 2}{20})}$ that means if we move away, the effect on this function is far less.  
  
    NOW IF WE LOOK AT THIS GRAPH AND WE IGNORE THE GREEN GRAPH AND TAKE THE 0 AS 1 AND IGNORE THE REST NUMBERS AND IGNORE THE $\mu$, WE CAN SEE THE SPREAD OF THE GRAPH INCREASES AS THE $\sigma$ INCREASES.  
For a simple reasion, we are dividing the denominator by a large number, if we divide the denominator by a large number, we have to change numerator a lot more to get to the same value so that spread is much bigger.  
So spread depends upon the value of $\sigma^2$.  
And effect will be exactly opposite when we consider this to be $e ^{-\lambda||x - l_{i}||^2}$.  
If this is the case we can say if $\lambda$ is large, that means $\sigma^2$ is smaller, that means we get lower spread, if $\lambda$ is very small, its reducing the effect of $||x - l_{i}||$ so the spread will be a lot more.  
What is the meaning of this?  
Lets say we have $l_{1}$, $l_{2}$, $l_{3}$, if $\sigma^2$ is large, that means $l_{1}$ will have  effect on far far away areas as well. If $\sigma^2$ is very small, each point will have effect only on the vicinity and they will not have effect on further away points. This will lead to overfitting.  
Because we care abt points which is nearest and not caring about other points.  
If we make $\sigma^2$ to be very small, it will lead to overfitting. (good training result)  
If we make $\sigma^2$ to be large, it will lead to underfitting.  
Somewhere in between will have ideal value of $\sigma^2$ where we get optimal result.
