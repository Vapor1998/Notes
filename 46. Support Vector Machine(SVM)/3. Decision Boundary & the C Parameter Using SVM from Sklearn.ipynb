### Effects of C parameter 
Lets say we have some data.  
We have two classes seprated well. 

Now we have multiple options for decision boundary, one is a straight line between them. other is little tilted on the left side.  
The tilted one will be the best(green) because it keeps the good gaps in the decision boundary.  
So why exactly the SVM try to get here.  
So we need to understand there are two parts in the cost function. First where we are trying to minimise the cost associated with the training data labels and then we have $\frac{1}{2}\sum \theta^2$ for all the thetas for regularization parameter and the cost. So how will we actually get to tilted decision boundary instead of straight?  
What we are trying to do is $\theta^T x$ $>=$ 1 when y = 1. SO we are trying to increase the value above a certain number.  
Actually, to calculate $\theta^T x$, there is another way of doing this.  
If we just look at the distance from the decision boundary of any training datapoint, for eg, the distance from the straight decision boundary is very less. Lets say the distance is d. So to find $\theta^T x$, it is actually same as   $\theta^T x$ = $ d||\theta||$, which is actually $\sqrt(\sum \theta^2)$.  
So this opertion is trying to find the length of $\theta$ vector.  
SO if the distance is very less, in that case the $\theta $ has to be much larger.  
If we agree on the fact that if we just find the distance then we can actually find $\theta^T x$ by something like above.  
And in case we are not maintaining the goood distance of training datapoints $\theta$ will have to be much larger for $\theta^T x$ to be >= 1.  
Because we want this value to be less, as said by the regularization parameter we want $ ||\theta||$ to be less, thats exactly why we want your d to be more so that this multiplication $ d||\theta||$ still stays $ d||\theta||$ >= 1.  
And thats exactly the case for <= -1.  
So we want to maintain this distance as much as possiblebecause higher the distance, the lower $\theta$ values you can actually try and reach.  
SO thats exactly why we want to find tilted decision boundary instead of the straight.  
This was the simple case , things can be complicated.  
So lets say we have the similar case where we reached the tilted boundary if we had this data we can have one cross data in the dot data.  
We can either maintain the large gap and have one error in our training label, or we can end up with decision boundary like where there is another tilted boundary in the opposite side where it takes the one cross inside it and seprate both the classes.But because this distance is very less, we will have to have larger values of $\theta$  which will lead to more error on the regularisation part.  
So $$ cost = c (y_{i} Cost_{1}(\theta x) + (1 - y_{i}) Cost_{0}(\theta x) + \frac{1}{2} \sum \theta^2)$$

SO lets call the first tilt as d1 and the other tilt which was on opposite side is d2.  
$\frac{1}{2} \sum \theta^2)$ will be less because we are at distance from all training datapoints so because the distance is more, we can choose smaller value of $\theta$ but in case of d2, we are very close to our datapoint so $\frac{1}{2} \sum \theta^2)$ will be higher so when d1 lets call it $  c (y_{i} Cost_{1}(\theta x) + (1 - y_{i}) Cost_{0}(\theta x)$ as $p1$ of the cost and $\frac{1}{2} \sum \theta^2)$ as $p2$ of the cost.  
So in case of d1 we are making error so $p1$ will be high and $p2$ will be low.  
In case of d2, our $p1$ is low because we are not making any error and $p2$ will be higher because the distance is less, we will end up getting higher value of $\theta$.  
Now whether we choose d1 or d2, depends upon the parameter c.  
if c is very large, we want to make sure $p1$ is less. If c is large, we will end up with d2 as decision boundary.  
If c is small, we will end up as d1 as our decision boundary.  
So depending upon value of c, you will either overfit, or you will underfit if we take c very small, but we need to find the right balance and find the optimal value of c. So higher values of c will lead to overfitting and lower value of c leads to underfitting.
