we have our cost to be $log(h_{\theta}x)$, this was our cost in case of Logistic Regression.  
which is equal to $ log (\frac{1}{1+e^{-\theta x}})$.  
In case of y = 0, our cost was $log(1- h_{\theta}(x))$  which is $ log (1 - \frac{1}{1+e^{-\theta x}})$.  
Lets try to plot this cost when y = 1.  
Lets take $\theta x$ to be the x axis.  
As the $\theta x$, cost decreases. Whar happens when $\theta x$ is 0.?

What do we want is we want not just  $\theta x $ >= 0, but $\theta x $ >= 1, so we want enough margin, and if we have enough margin,  we dont care abt how much more margin do we have, so what we are gonna do is we are going to change this cost function.

And at the value of 1 on x axis, we will make the cost to be 0.  
And for otherwise, we are actually going to change the cost function otherwise as well, we are going to make it linear.  
SO our cost function will be linear when $\theta x $ <= 1, and after $\theta x $ >= 1, we will make it 0.  
Similarly here, if we are crossing a particular threshhold, which is 1 in our case, we are making the cost to be 0, otherwise we are going to make it something linear.  
So we are changing the cost function into two separate line segments one segment when $\theta x $ starts from 1, in case of y = 1, and its basically 0 line.  
And the other line which is basically, increases as your $\theta x $ decreases, for y = 1 and vice versa for y = 0.  
So basically we have changed our cost function and its basically a combination of below two graphs.  




To write it formally, in case of logistic regression, our cost function was $$ cost = y_{i} log(h_{\theta}(x)) + (1 - y_{i})log(1 - h_{\theta}x) + \frac{\lambda}{2} \sum \theta^2$$. This is the cost function in case of Logistic Regression.  
For SVM, 
$$ cost = y_{i} Cost_{1}(\theta x) + (1 - y_{i}) Cost_{0}(\theta x) + \frac{\lambda}{2} \sum \theta^2$$  where $Cost_{1}$ and $Cost_{0}$ is the same function from the graph when $\theta x$ is 0 and 1 respectively.  
Generally in case of SVM, we tend to get rid of the $\lambda$ and instead we do $$ cost = \frac {1}{\lambda}(y_{i} Cost_{1}(\theta x) + (1 - y_{i}) Cost_{0}(\theta x) + \frac{1}{2}\sum \theta^2)$$, now for $\frac{1}{\lambda}$ we use c instead for this.  
so cost function looks like $$ cost = c (y_{i} Cost_{1}(\theta x) + (1 - y_{i}) Cost_{0}(\theta x) + \frac{1}{2} \sum \theta^2)$$

SO we will be minimising this $\theta $ values or we will be minising the cost function to get the optimal $\theta$ values
