```python
import docx2txt
import re 
import numpy as np
import pandas as pd

mytext = docx2txt.process(r"C:\Users\HASHIM RAZI\Downloads\nato handbook (1).docx")

```

```python
def textretrieve(mytext):
    text = mytext.split(' ')
    
    #arr = ["Part I", "Part II",  "PART III", "PART IV", "PART V", "PART VI", "PART VII", "PART VIII", "PART IX", "PART X", "The Strategic Commanders have a staff representative on each board"]
    arr = ['An introduction to the Alliance',
           'Policy and decision-making', 
           'NATO’s civilian and military structures',
           "The Alliance's role in peacekeeping and peace-support operations",
           'Combating new threats and developing new capabilities',
           'The opening up of the Alliance to new member countries',
           'Partnership and cooperation', "NATO-EU RELATIONS", 
           "The wider institutional framework for security",
           "Programmes, Activities, Organisations and Agencies",
           "The Strategic Commanders have a staff representative on each board"]
    
    count = 0
    arr1 = []

    while count != 10:
        nextcount = count
        start = arr[count]
        end = arr[nextcount+1]

        li = mytext.find(start)
        mi = mytext.find(end)

        
        # extracting text
        if li != -1 and mi != -1:
            txt = mytext[li + len(start):mi]
        #print(extracted_text)
        else:
            print("Markers not found in the paragraph.")

        x = []
        for i in txt.split('\n\n'):
            x.append(i)


        # appending text
        s = ' '
        for i in x:
            if len(i) >= 3:
                y = re.findall('\w.+', i)
                z = y[0]
                if ("CHAPTER" in z) or ("Chapter"  in z): # exclude the word chapter since we have it.
                    continue

                if ':' in z[-1]:
                    z.replace(':', '')
                    s += z
                    continue

                if ';' in z[-1]:
                    z = z.replace(';',',')
                    s += z
                    continue

                if '.' not in z[-1]:
                    s += z

                else:
                    if len(s) > 1:
                        s+=z
                        arr1.append(s)
                    else:
                        arr1.append(z)

                    s = ' '

        #print(arr1)
        count+=1
        
    return arr1
```
