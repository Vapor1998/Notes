Lets go back to Logistic Regression and talk abt the problem we faced there.  
LR is basically we used sigmoid function $\frac{1}{1 + e^-{\theta x}}$ to find out which class a particular input belongs to.  
And we were learning this theta parameters.  
Now what we realized after learning logistic regression is even though this function is non-linear function, but we could reach only Linear decision boundaries using this.  
That was basically because we put a limit that $if \frac{1}{1 + e^-{\theta x}} > 0.5$, Im going to predict something and if $if \frac{1}{1 + e^-{\theta x}} < 0.5$ I want to predict something else.  
And then we realized it became dependent on $\theta x$, where $ \theta x  >= 0$ we started to predict something and if $ \theta x  < 0$ we started to predict something.  
So it became a linear decision boundary even though the function is not Linear.  
So we decided in LR for non-linear decision boundaries, lets say if we have 5 features say $f_1$, $f_2$, $f_3$, $f_4$, $f_5$, 
what we can do is we can add new features, which can be $f_1^2$, $f_2^2$, $f_3^2$, $f_4^2$, $f_5^2$ and any kind of features that we want to add. And then LR will find the linear function w.r.t this new featureset which actually is non-linear in the original dataset that we have. This is the one thing that we did to find the non-linear decision boundaries, in case of LR.  
Was that a good solution? It is a pretty decent solution the only problem with this solution is that we need to figure out what kind of features we need to add.  
What if the Decision boundary is something like tangent or sine of the features. So we dont know what features to add.  
That is a decision that we were supposed to take.  
What we want is the Algorithm that we are using should figure out the kind of decision boundaries it needs to come up with.  
We tried it by using SVM with Non-Linear Kernels but again there were also like of choosing what kind of Non-linear features we want to use again it was competitionally very expensive.
