```python
import pandas as pd
import numpy as np
```

```python
df_training=np.genfromtxt('0000000000002417_training_boston_x_y_train.csv',delimiter=',')
#df_training.head()
x_train=df_training[:,:-1]
y_train=df_training[:,-1]
x_train
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
x_train[:,:]=SS.fit_transform(x_train[:,:])
x_train
```

```python
def cost(x,y, m, c):
    total_cost = 0
    M = len(x)
    for i in range(M):
        x1 = x[i, :]
        y1 = y[i]
        total_cost += (1/M)*((y1 - (m*x1).sum() - c)**2)
    return total_cost
```

```python
def mc_calc(x,y, learning_rate, m , c):
    m_slope = list(0 for ni in range(x.shape[1]))
    c_slope = 0
    rows = len(x)
    a=0
    for i in range(rows):
        x1 = x[i, :]
        y1 = y[i]
        a=m*x1
        #print(a.shape)
        for j in range(x1.shape[0]):
            m_slope[j]+=(-2/rows)* (y1 - (a.sum(axis=0)) - c)*x1[j]
        c_slope += (-2/rows)* (y1 - (a.sum(axis=0)) - c)
    m_slope=np.array(m_slope)
    new_m = m - (learning_rate*m_slope)
    new_c = c - (learning_rate*c_slope)
    return new_m, new_c
```

```python
def gradient_descent(x,y, learning_rate, num_iterations):
    m=[0 for i in range(x.shape[1])] 
    c = 0
    for i in range(num_iterations):
        m, c = mc_calc(x,y, learning_rate, m , c)
        print(i, " Cost: ", cost(x,y, m, c))
    return m, c
```

```python
learning_rate = 0.1
num_iterations = 500
m, c = gradient_descent(x_train,y_train, learning_rate, num_iterations)
print(m, c)
```

```python
df_test=np.genfromtxt('0000000000002417_test_boston_x_test.csv',delimiter=',')
df_test
```

```python
df_test[:,:]=SS.fit_transform(df_test[:,:])
df_test
```

```python
y_pred=np.array((m*df_test).sum(axis=1))+c
y_pred
```
