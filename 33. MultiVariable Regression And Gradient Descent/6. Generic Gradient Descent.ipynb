```python
def step_gradient(X,Y, learning_rate, m ):
    M = len(X)

    m_slope = np.array([0 for i in range(len(X[0]))])
    
    for i in range(M):  ## This is the loop which iterates through all the values of X.
        
        x = X[i]   
        y = Y[i] 
        
        for j in range(len(m_slope)):   ## Running this loop len(m_slope) because we want to update all the values of m'
        
            m_slope[j] += (-2/M) * (y - ((m *x).sum()))*x[j]    ## calculating m slope inside loop
    
    new_m = m -  (learning_rate * m_slope)
    return new_m
```

```python
def cost(X, Y, m):
    ## after defining the cost(), we reduce iterations from 1000 to 10, and after every iteration, we print the cost.
    ## in gd().
    
    total_cost = 0
    M = len(X)
    for i in range(M):
        x = X[i]
        y = Y[i]
        total_cost += (2/M) * (Y[i]- sum(m*X[i]) )
    return total_cost
```

```python
def gd(X, Y, num_iterations, learning_rate):
    m=np.array([0 for i in range(len(X[0]))])
    for i in range(num_iterations):
        m=step_gradient(X, Y, learning_rate, m)
        
        print(i, 'Cost: ', cost(X,Y, m))
    return m
```
