```python
from selenium import webdriver

driver = webdriver.Chrome(executable_path = r"C:\Users\hashi\Downloads\chromedriver_win32 (1)\chromedriver.exe")  
## To get the path of webdriver, we need to go to the webdriver file we extracted, then press shift and rightclick,
## then we get the option of copy as path.


## For more, u must add r before the driver path and install latest chromedriver and then pass its path.
driver

## After running this, new empty web browser will open.
```

```python
## Lets say we want to visit coding ninjas in the web driver.
## We use get() and inside it, we pass website link we want to visit.

driver.get('https://codingninjas.in/')
```

```python
## Extracting data from this webdriver
## To extract title of this webpage, we use title

driver.title
```

```python
## To click the back button of this web browser, we use back()

driver.back()

## After clicking this, it will go to the previous page like we use to go to previous pages on website.
```

```python
## To click the forward button, we use forward()
driver.forward()

## After clicking this, it will go to one page forward like we do in web browsers.

```

```python
## To take screenshot, we use save_screenshot(), and we pass the name.png of the img file

driver.save_screenshot('testfile.png')
```

```python
## To get the html content of the page, we use page_source

driver.page_source
```

```python
## Lets say if we want to find an element by link
## Lets say if we want to find a link of which we know the text and we want to visit the link by clicking on it,
## we find the link by using find_element_by_link_text('elementname')

driver.find_element_by_link_text('Events')

## The below o/p indicates we have found the element and link, now to click on it we add this in variable,
## and then call this variable with click()
```

```python
ev_btn = driver.find_element_by_link_text('Events')
ev_btn.click()

## As we can see, we have visited the new page which we found.
```

```python
## Lets say if we dont know the exact text of which hyperlink is created, we just know the partial part, we use
## find_element_by_partial_link_text('elementname')

b_btn = driver.find_element_by_partial_link_text('Bl')
b_btn.click()

## As we can see, we wanted to go to the blog but wedidnt knew the full element, it visited the page perfectly.
```

```python
## What if we give the element that didnt exist

ev_btn = driver.find_element_by_link_text('Eventswe')
ev_btn.click()
```
